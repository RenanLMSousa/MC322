# Exercício do Empréstimo

Em um financiamento com juros compostos e número de parcelas fixas parte-se dos seguintes parâmetros:
* `S` - valor da primeira parcela
* `N` - número de parcelas
* `J` - percentual de juros mensal

A primeira parcela a ser paga do financiamento é sempre igual a `S`. A partir daí é feita uma atualização mensal da parcela, em que cada nova parcela é calculada a partir da parcela do mês anterior, conforme a fórmula:

> #### Parcela<sub>mês</sub> = Parcela<sub>mês-1</sub> * (1 + `J` / 100)

O financiamento encerra quando as `N` parcelas são pagas.

Exemplo:
* `S`: 200
* `N`: 5
* `J`: 1%

Parcelas do financiamento:
`200`; `202`; `204.02`; `206.06`; `208.12`

## Exercício Parte 1 - Escrevendo um Módulo

Dado o problema descrito, escreva um programa que calcule as parcelas de um empréstimo para os seguintes valores:
* `S`: 200
* `N`: 5
* `J`: 1%

```c
#include <stdio.h>

#define VAL_PRIM_PARCELA 200
#define N_PARCELAS 5;
#define JUROS 1

int main(){
    
    float S = VAL_PRIM_PARCELA;
    int N = N_PARCELAS;
    float J = JUROS;
    
    for(int i =0;i<N;i++){
        printf("%.2f\n",S);
        S *= (1+J/100);
        
    }
    
    return 0;
}
```

## Exercício Parte 2 - Escrevendo um Módulo

Reescreva o código acima de forma que seu programa faça uso de uma função que seja responsável pelo cálculo de uma parcela X do empréstimo. Utilize as boas práticas de modularização que você aprendeu, evitando dependências do módulo com o programa principal.

```c
#include <stdio.h>

float valor_parcela_x(float S,int N,float J){
    //Retorna o valor da parcela N
    float valor_parcela=S;
    for(int i = 1;i<N;i++){
        valor_parcela *= (1+J/100);
    }
    return valor_parcela;
    
}

int main(){
    
    int N = 5;
    float S= 200;
    float J=1;
    
    for(int i=1;i<=N;i++){
        printf("%.2f\n",valor_parcela_x(S,i,J));
    }
  
    
    return 0;
}
```

* evite informar recorrentemente como parâmetros dados sobre as características do empréstimo;
* evite usar informações de controle (e.g., parcela corrente) como parâmetro.

```c
#include <stdio.h>
#include <stdlib.h>

float valor_proxima_parcela(float *  S,float J){
    //Retorna o valor da próxima parcela
    
    float prox_parcela = *S;
    *S *= (1+J/100);
    return prox_parcela;
    
}


int main(){
    float * S = malloc(sizeof(float));
    *S=200;
    float J =1;
    int N=5;
    
    for(int i=0;i<N;i++){
        printf("%.2f\n",valor_proxima_parcela(S,J));
    }
    
    return 0;
}
```

## Exercício Parte 4 - Minimizando os Parâmetros

Partindo da Solução da Parte 2, modifique a forma como você representa o empréstimo, de forma que a função que calcula a próxima parcela receba um único parâmetro.

```c
#include <stdio.h>
#include <stdlib.h>

typedef struct emprestimo{
    
    int N;
    int Natual;
    float S;
    float J;
    
}Emprestimo;

Emprestimo * criar_emprestimo(int N,float S, float J){
    //Cria um empréstimo com os valores fornecidos
    Emprestimo * emprestimo = malloc(sizeof(Emprestimo));
    emprestimo->N=N;
    emprestimo->Natual=1; //Ainda a pagar
    emprestimo->S=S;
    emprestimo->J=J;

    return emprestimo;
    
}

float valor_parcela_x(float S,int N,float J){
    //Calcula o valor da parcela N
    float valor_parcela=S;
    for(int i = 1;i<N;i++){
        valor_parcela *= (1+J/100);
    }
    return valor_parcela;
    
}

float valor_proxima_parcela(Emprestimo * emprestimo){
    //Calcula o valor da próxima parcela
    float proxima_parcela = valor_parcela_x(emprestimo->S,emprestimo->Natual,emprestimo->J);
    emprestimo->Natual +=1;

}


int main(){
    Emprestimo * emprestimo = criar_emprestimo(5,200,1);
    for(int i=0;i<emprestimo->N;i++){
        printf("%.2f\n",valor_proxima_parcela(emprestimo));
    }
    
    free(emprestimo);
    return 0;
}
```

## Exercício Parte 5 - Múltiplos Empréstimos

Considere que há múltiplos empréstimos que podem ser controlados em paralelo. O seu programa deve ser capaz de informar as parcelas de mais de um empréstimo, mantendo um controle para cada empréstimo separadamente. Você deve decidir qual das soluções tomará como ponto de partida, se deve modificar o programa principal, a função ou ambos para atender esse requisito da melhor forma possível.

Por exemplo, suponha os seguintes dois empréstimos em paralelo:

### Empréstimo 1
* `S`: 200
* `N`: 5
* `J`: 1%

### Empréstimo 2
* `S`: 500
* `N`: 7
* `J`: 2%

A saída esperada é:
~~~
Emprestimo 1: parcela 1 eh 200.00
Emprestimo 2: parcela 1 eh 500.00
Emprestimo 1: parcela 2 eh 202.00
Emprestimo 2: parcela 2 eh 510.00
Emprestimo 1: parcela 3 eh 204.02
Emprestimo 2: parcela 3 eh 520.20
Emprestimo 1: parcela 4 eh 206.06
Emprestimo 2: parcela 4 eh 530.60
Emprestimo 1: parcela 5 eh 208.12
Emprestimo 2: parcela 5 eh 541.22
Emprestimo 2: parcela 6 eh 552.04
Emprestimo 2: parcela 7 eh 563.08
~~~

O exemplo mostra dois empréstimos, mas a estrutura deve ser genérica o suficiente para controlar N empréstimos em paralelo.

```c
#include <stdio.h>
#include <stdlib.h>



typedef struct emprestimo{
    
    int N;
    int Natual;
    float S;
    float J;
    
}Emprestimo;

typedef struct no{
    
    struct no * proximo;
    Emprestimo * emprestimo;
    int ordem;
    
}No;

typedef struct lista{
    
    No * cabeca;
    int n_elementos;
    int maior_emprestimo;
    
}Lista;

Lista * criar_lista(){
    //Cria uma lista vazia
    Lista * lista = malloc(sizeof(Lista));
    lista-> cabeca = NULL;
    lista->n_elementos =0;
    lista->maior_emprestimo=0;
    return lista;
    
}

No *  adicionar_lista(No * no, Emprestimo * emprestimo,int n){
    //Adiciona um empréstimo à lista
    No *novo;
    novo = malloc(sizeof(No));
    novo->emprestimo = emprestimo;
    novo->proximo = no;
    novo->ordem = n;
    return novo;
}

Emprestimo * criar_emprestimo(int N,float S, float J){
    //Cria um empréstimo com as informações fornecidas
    Emprestimo * emprestimo = malloc(sizeof(Emprestimo));
    emprestimo->N=N;
    emprestimo->Natual=1; //Ainda a pagar
    emprestimo->S=S;
    emprestimo->J=J;

    return emprestimo;
    
}


float valor_parcela_x(float S,int N,float J){
    //Calcula o valor da parcela N
    float valor_parcela=S;
    for(int i = 1;i<N;i++){
        valor_parcela *= (1+J/100);
    }
    return valor_parcela;
    
}

float valor_proxima_parcela(Emprestimo * emprestimo){
    //Retorna o valor da próxima parcela
    
    float proxima_parcela = valor_parcela_x(emprestimo->S,emprestimo->Natual,emprestimo->J);
    emprestimo->Natual +=1;

}

void adicionar_emprestimo(Lista * lista,int N,float S, float J){
//Adiciona um novo empréstimo a uma lista existente
    
    Emprestimo * novo = criar_emprestimo(N,S,J);
    lista->n_elementos++;
    lista->cabeca  = adicionar_lista(lista->cabeca,novo,lista->n_elementos);
    if(N>lista->maior_emprestimo){
        lista->maior_emprestimo = N;
    }


}

void imprimir_lista(No * no,int parcela){
//Imprime os próximos empréstimos da lista que ainda não foram quitados
    if(no !=NULL){
     imprimir_lista(no->proximo,parcela);
        if(parcela<=no->emprestimo->N){
            printf("Emprestimo %d: parcela %d eh %.2f\n",no->ordem,parcela,valor_proxima_parcela(no->emprestimo));
        }


    }

    
}

void imprimir_saida(Lista * lista){
//Imprime todos os empréstimos
    for(int i =1;i<=lista->maior_emprestimo;i++){
        
        imprimir_lista(lista->cabeca,i);
    }

}


int main(){
    
    Lista * lista = criar_lista();
    adicionar_emprestimo(lista,5,200,1);
    adicionar_emprestimo(lista,7,500,2);
    imprimir_saida(lista);



    
    return 0;
}
```
